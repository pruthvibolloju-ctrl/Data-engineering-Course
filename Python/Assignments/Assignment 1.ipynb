## For loop Examples

```python
for i in range(1, 6):
    print("*" * i)
```

I used a for loop to print a right-angled triangle pattern.
The loop starts with i = 1, it prints * one time.
when i = 2, it prints * two times
This continues until the condition is satisfied, i.e., i = 5. It will not go to 6 because it will not take the upper limit value.

```python
for i in range(5, 0, -1):
    print("*" * i)
```

I used a loop to print an inverted star triangle.
In this the loop starts at 5 and prints five stars and then decreases i by one each time.
and this loop will run until i becomes 0 and it will stop and we get the output.

```python
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

I used a loop with in a loop to print Number Triangle Pattern.
first the outer loop is used to control rows, then it will go into inner loop where it will take i as 1 then i will print 1 and then i value is incremented every time. and the output is obtained. 

```python
rows = 5
for i in range(1, rows + 1):
    print(" " * (rows - i) + "*" * (2 * i - 1))
```

I used a for loop to create a star pyramid pattern.

For each row: First it prints spaces (rows - i), then prints stars (2 * i - 1). The formula (2*i - 1) always gives odd numbers, i.e., 1, 3, 5, 7, 9. That keeps the pyramid symmetrical. It will run until the condition is satisfied and the loop stops.

```python
num = 1

for i in range(1, 6):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```

I used a nested for loop to create Floyd's triangle.
num starts from 1. Every time the inner loop runs, it prints num. Then num increases by 1. Numbers continue without resetting.

## while loop examples

```python
i = 1

while i <= 5:
    print("*" * i)
    i += 1
```

In this program, I used a while loop to print a right-angled triangle pattern using stars. First, I gave the starting value i = 1. The loop runs as long as i is less than or equal to 5. Inside the loop, it prints stars equal to the value of i. After printing, I increase the value of i by 1. So in the first iteration, it prints 1 star, then 2 stars, then 3, and so on. When i becomes 6, the condition becomes false and the loop stops. That is how the triangle shape is formed.

```python
i = 5

while i >= 1:
    print("*" * i)
    i -= 1
```

In this program, I printed the reverse of the triangle pattern. I started with i = 5. The loop runs while i is greater than or equal to 1. Inside the loop, it prints stars equal to i. After printing, I decrease the value of i by 1. So first it prints 5 stars, then 4, then 3, and continues decreasing. When i becomes 0, the condition becomes false and the loop stops. This creates an inverted triangle shape.

```python
i = 1

while i <= 5:
    j = 1
    while j <= i:
        print(j, end=" ")
        j += 1
    print()
    i += 1
```

In this program, I used two while loops (nested loops). The outer loop controls the number of rows, and the inner loop prints numbers in each row. The outer loop starts with i = 1 and runs until i is 5. Inside it, the inner loop starts with j = 1 and prints numbers from 1 up to the value of i. After finishing one row, I move to the next line and increase i. Because of this, each row prints one more number than the previous row, forming a number triangle.

```python
rows = 5
i = 1

while i <= rows:
    print(" " * (rows - i) + "*" * (2 * i - 1))
    i += 1
```

In this program, I printed a pyramid shape using stars. First, I set the number of rows and started i = 1. In each iteration, I printed some spaces before the stars to keep the pattern in the center. The number of spaces decreases as i increases. Then I printed stars using the formula (2 * i - 1) so that the number of stars increases in odd numbers like 1, 3, 5, 7, and 9. This makes the pyramid symmetrical. The loop continues until all rows are printed.

```python
i = 1
num = 1

while i <= 5:
    j = 1
    while j <= i:
        print(num, end=" ")
        num += 1
        j += 1
    print()
    i += 1
```

In this program, I printed numbers continuously without restarting from 1 in each row. I used two while loops. The outer loop controls the rows, and the inner loop prints numbers in each row. I created a variable num starting from 1. Every time a number is printed, num increases by 1. Because of this, the numbers continue in sequence across all rows. The loop stops when the outer loop condition becomes false.
