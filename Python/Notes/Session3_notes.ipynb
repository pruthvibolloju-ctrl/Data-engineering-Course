# Session 2: Strings & List Objects in Python

## 1. Strings in Python

Strings are sequences of characters enclosed in single ' ' or double " " quotes.

```python
# Creating strings
name = "Pruthvi"
greeting = 'Hello World'
print(name, greeting)
```

Variables store text. Strings can be single or double quotes. print() shows them.

### String Operations

```python
# Concatenation
first = "Hello"
second = "World"
full = first + " " + second
print(full)
```

this + joins strings; add space manually if needed.

```python
# Repetition
text = "Ha! "
print(text * 3)
```

* repeats the string multiple times.

```python
# Indexing
word = "Python"
print(word[0], word[3], word[-1])
```

Index 0 is first character, -1 is last character.

```python
# Slicing
text = "Programming"
print(text[0:6])
print(text[3:])
print(text[:6])
```

[start:end] extracts characters from start index up to end-1. Omitting start or end uses default positions.

```python
# String Methods
name = "python"
print(name.upper()) #to print uppercase
print(name.lower()) #to print lower case
print(name.capitalize()) #to captalize first letter
print("  hello  ".strip()) #removes extra spaces
```

Methods manipulate string appearance or remove spaces.

```python
# Special String Checks
s = "123"
print(s.isdigit()) 
print(s.isalpha()) 
```

isdigit() checks for numbers, and isalpha() for letters.

```python
# Split and Join
sentence = "Python is fun"
words = sentence.split()
print(words)
joined = "-".join(words)
print(joined)
```

split() converts a string to a list; join() combines a list into a string.

## 2. List Objects

```python
# Creating Lists
numbers = [1, 2, 3, 4]
fruits = ["apple", "banana", "cherry"]
mixed = [1, "hello", 3.5]
print(numbers, fruits, mixed)
```

Lists store multiple items of any type.

```python
# Indexing and Slicing
fruits = ["apple", "banana", "cherry"]
print(fruits[0], fruits[-1])
numbers = [10, 20, 30, 40, 50]
print(numbers[1:4], numbers[:3], numbers[2:])
```

Access single elements with index; slices extract multiple elements.

```python
# List Methods
fruits = ["apple", "banana"]
fruits.append("orange")
fruits.insert(1, "grape")
fruits.remove("banana")
fruits.pop()
numbers = [3,1,2]
numbers.sort()
numbers.reverse()
print(fruits, numbers)
```

Methods allow adding, removing, sorting, and reversing list items.
append(x)       Add x at the end
insert(i, x)	Add x at index i
remove(x)	    Remove first occurrence of x
pop()	        Remove last element	fruits
sort()	        Sort list
reverse()	    Reverse list
count(x)	    Count occurrences
index(x)	    Index of first occurrence

```python
# List Operations
list1 = [1,2]
list2 = [3,4]
combined = list1 + list2
print(combined)  # Concatenation
nums = [0,1]
print(nums*3)   # Repetition
```

+ merges lists, * repeats list contents.

```python
# Nested Lists
matrix = [[1,2,3],[4,5,6],[7,8,9]]
print(matrix[0], matrix[0][1])
```

Lists can contain lists. Access elements using multiple indices.

```python
# List Comprehension
squares = [x**2 for x in range(1,6)]
print(squares)
```

Creates a new list by applying an expression to each item.

```python
# Strings and Lists Together
sentence = "Python is fun"
words = sentence.split()
joined = " ".join(words)
print(joined)
```

split() converts string to list; join() converts list back to string.
