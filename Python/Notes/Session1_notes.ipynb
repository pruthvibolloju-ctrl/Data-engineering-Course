## Python Introduction
Python is a high-level, dynamically typed, multiparadigm programming language. It is simple to read and write and requires fewer lines of code compared to many other languages.

```python
print("Hello, Python!")
```

### Python vs Java Comparison
Python code is shorter and simpler compared to Java.

```python
print("Hello World")
```

```python
x = 5
name = "John"
```

```python
a = 10
print(type(a))

a = "Hello"
print(type(a))
```

## Strong Typing Example

Even though Python is flexible, it does not allow mixing incompatible types automatically.

Incorrect Example (Will cause error):

```python
a = 10
print("Number is " + a)  # Uncommenting will cause error
```

This will give an error because we cannot directly add a string and an integer.

Correct Example:

```python
a = 10
print("Number is " + str(a))
```

Here, we convert the integer into a string using str().

## Simple Expressions

```python
a = 5
b = 2
print(a + b)
print(a * b)
```

### Boolean Expressions

```python
a = 10
b = 5
print(a > 5 and b < 10)
print(a > 15 or b < 10)
print(not(a > b))
```

## Branching (if / elif / else)

Branching allows the program to make decision

```python
age = 18

if age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible")
```

```python
num = -5

if num > 0:
    print("Positive number")
elif num < 0:
    print("Negative number")
else:
    print("Zero")
```

```python
marks = 78

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```
