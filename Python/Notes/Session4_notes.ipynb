## Tuple
### What is a Tuple?
A tuple is:
- Ordered (keeps position)
- Immutable (cannot be changed)
- Written using round brackets ()


```python
# Creating a Tuple
my_tuple = (10, 20, 30)
```

```python
# Accessing Elements
numbers = (5, 10, 15, 20)
print(numbers[0])     # First element
print(numbers[2])     # Third element
print(numbers[-1])    # Last element
```

```python
# Tuple Packing and Unpacking
data = 1, 2, 3
a, b, c = data
print(a, b, c)
```

```python
# Nested Tuple
nested = ((1, 2), (3, 4))
print(nested[1][0])
```

## Set
### What is a Set?
A set:
- Does not allow duplicates
- Is unordered
- Written using {}


```python
# Creating a Set
numbers = {1, 2, 2, 3, 3}
print(numbers)   # Duplicates removed
```

```python
# Adding & Removing
my_set = {1, 2}
my_set.add(3)
print(my_set)

my_set.remove(2)
print(my_set)
```

```python
# Set Operations
a = {1, 2, 3}
b = {3, 4, 5}

print("Union:", a | b)
print("Intersection:", a & b)
print("Difference:", a - b)
```

```python
s = {1, 1, 2, 3}
print(s)
```

```python
s.add(4)
print(s)
```

```python
x = {1, 2, 3}
y = {2, 3, 4}
print(x & y)
```

```python
lst = [1, 2, 2, 3]
print(set(lst))
```

## Dictionary
### What is a Dictionary?
- Stores data in key-value pairs
- Written using {}


```python
student = {
    "name": "John",
    "age": 20,
    "grade": "A"
}
```

```python
print(student["name"])  # John #access values
```

```python
print(student.get("age"))  # 20 #using .get()
```

```python
student["city"] = "New York" #adding new key value
```

```python
student["age"] = 21 #updating value
```

```python
student.pop("grade") # removing item using .pop()
```

```python
for key in student:
    print(key)  #loop through keys
```

```python
for value in student.values():
    print(value)  #loop through values
```

```python
for key, value in student.items():
    print(key, ":", value) # loop through both key and value
```

```python
students = {
    "student1": {"name": "Alice", "age": 20},
    "student2": {"name": "Bob", "age": 22}
}

print(students["student1"]["name"])  # Alice      #nested dictionary
```

## Functions
### What is a Function?
A reusable block of code defined using def.

```python
#syntax
def function_name(parameters):
    # code
    return value
```

```python
# Simple Function
def greet():
    print("Hello!")

greet()
```

```python
# Function with Parameters
def greet_name(name):
    print("Hello", name)

greet_name("Alice")
```

```python
# Function with Return
def add(a, b):
    return a + b

print(add(5, 3))
```

```python
# Multiple Return
def calculate(a, b):
    return a + b, a - b

sum_result, diff_result = calculate(10, 5)
print(sum_result, diff_result)
```

```python
#default parameters
def greet(name="Guest"):
    print("Hello", name)

greet()        # Hello Guest
greet("John")  # Hello John
```

```python
#keyword arguments
def student_info(name, age):
    print("Name:", name)
    print("Age:", age)

student_info(age=20, name="Alice")
```

```python
#function inside function
def outer():
    def inner():
        print("Inside inner function")
    inner()

outer()
```

```python
# Even Check
def is_even(n):
    return n % 2 == 0

print(is_even(4))
```

```python
# Factorial
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial(5))
```
