# Session 2: Python Loops

## For Loop

A for loop is used to iterate over a sequence like
- List
- Tuple
- String
- range()

```python
numbers = [2, 4, 6, 8]
for num in numbers:
    print(num)
```

### Find Sum of List

```python
numbers = [5, 10, 15]
total = 0

for num in numbers:
    total += num

print('Total is:', total)
```

## Using range()

range(start, stop, step) generates numbers.



```python
#Print numbers 1 to 5
for i in range(1, 6):
    print(i)
```

```python
#Print Even Numbers
for i in range(2, 11, 2):
    print(i)
```

## Loop Using Index

Sometimes we need index positions using len().

```python
fruits = ['apple', 'banana', 'mango']

for i in range(len(fruits)):
    print('Index:', i, 'Value:', fruits[i])
```

## Reverse a String Using Loop

```python
word = 'coding'
reverse_word = ''

for letter in word:
    reverse_word = letter + reverse_word

print(reverse_word)
```

## While Loop

A while loop runs as long as the condition is True.

```python
i = 1
while i <= 5:
    print(i)
    i += 1
```

```python
#Countdown example
count = 5
while count >= 0:
    print('Countdown:', count)
    count -= 1
print("Happy Birthday!")
```

## break Statement

```python
for letter in 'Python':
    if letter == 'h':
        break
    print(letter)

print('Loop ended')
```

## continue Statement

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

## Loop with else

The else block runs if loop finishes normally (without break).

```python
for i in range(3):
    print(i)
else:
    print('Loop finished successfully')
```

## User Input with Loop

```python
correct_name = 'pruthvi'

while True:
    user_input = input('Enter your name: ')
    if user_input == correct_name:
        print('Name matched!')
        break
    else:
        print('Try again')
```

## Practice questions

### 1. Multiplication Table of 5

```python
for i in range(1, 11):
    print('5 x', i, '=', 5*i)
```

### 2. Sum of First 10 Numbers using while loop

```python
i = 1
total = 0

while i <= 10:
    total += i
    i += 1

print('Sum is:', total)
```

### 3. Print Odd Numbers from 1 to 20

```python
for i in range(1, 21):
    if i % 2 != 0:
        print(i)
```

```python
#Number Guessing Game
secret_number = 7

while True:
    guess = int(input('Guess the number (1-10): '))
    
    if guess == secret_number:
        print('Congratulations! You guessed correctly.')
        break
    else:
        print('Wrong guess. Try again.')
```
